```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

prompt = """
Summarize the following text in two lines:
Large Language Models are AI systems trained on huge datasets to understand and generate human-like text.
"""

output = generator(prompt, max_length=80)
print(output[0]["generated_text"])
```

```python
from langchain.prompts import PromptTemplate
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

hf_pipeline = pipeline("text-generation", model="gpt2")
llm = HuggingFacePipeline(pipeline=hf_pipeline)

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple words."
)

prompt = template.format(topic="Prompt Engineering")
print(llm(prompt))
```
