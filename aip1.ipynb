```python
from transformers import pipeline
import time

# Load GPT-2
generator = pipeline("text-generation", model="gpt2")

prompt = "Artificial Intelligence will change the world by"

start = time.time()
output = generator(prompt, max_length=60, num_return_sequences=1)
end = time.time()

print("Prompt:", prompt)
print("Output:", output[0]["generated_text"])
print("Latency:", round(end - start, 2), "seconds")
```

```python
from transformers import pipeline
import time

generator = pipeline(
    "text-generation",
    model="EleutherAI/gpt-neo-1.3B"
)

prompt = "Artificial Intelligence will change the world by"

start = time.time()
output = generator(prompt, max_length=80, do_sample=True)
end = time.time()

print("Prompt:", prompt)
print("Output:", output[0]["generated_text"])
print("Latency:", round(end - start, 2), "seconds")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_id = "mistralai/Mistral-7B-Instruct-v0.1"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)

prompt = "Explain LLMs in simple words."

inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
outputs = model.generate(**inputs, max_new_tokens=100)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
